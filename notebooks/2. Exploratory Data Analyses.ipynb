## Choosing the State of Our Interest  - Oregon 

College Board's mission is to provide support and funding to every state within the United States to ensure students are prepared for college and for careers after school. Therefore it is necessary to find states which have persistently demonstrated low participation rates in both ACT and SAT. Since we only have 2017 and 2018 data, we can only search to identify states that have such low SAT and ACT participation rates in 2017 and in 2018. 

```python
# find the state which has low SAT and ACT participation rate in 2017


low_sat_act_part_17 = []
for i in low_sat_17['state']:
    for j in low_act_17['state']:
        if i in j:
            low_sat_act_part_17.append(i)

low_sat_act_part_17            
```

```python
# find the state which has low SAT and ACT participation rate in 2018

low_sat_act_part_18 = []

for i in low_sat_18['state']:
    for j in low_act_18['state']:
        if i in j:
            low_sat_act_part_18.append(i)

low_sat_act_part_18
```

**Oregon** is identified as the state with less than 50% participation rate both in SAT and in ACT in 2017 and 2018. This means that the total participation rate in college preparatory standardized testing in Oregon is below 100%, indicating that a lot of the graduating studenst in Oregon are left behind in college preparedness.
