### How Do State Requirements and Test Costs Affect the Relationship between SAT and ACT Participation Rates?

It would be interesting to see how the relationship between SAT and ACT participation rates differs as it's affected by state requirements of the exams and by whether the exams cost money to students. To investiate more, I've used
additional [data](https://www.princetonreview.com/college-advice/act-sat-state-requirements)
from 2016-2017 academic year published on the Princeton Review. This data was combined with our ACT/SAT 2017 data. 

### Note:
The Princeton Review data is from 2016-2017 academic year. Our SAT/ACT data is from 2017 and not given in the academic year interval. Two states, Colorado and Illinois had a policy change in 2017 - they made the SAT's mandatory which is reflected in the Princeton Review data. With mandatory SAT's, we expect nearly 100% participation rate for those states. We see the effect of mandating SAT in Colorado and Illunois as a jump from 11% to ~100% in Colorado, and 9% to ~100% in Illunois from 2017 to 2018 respectively. However, because the mandate of the test occurred in 2017, there was a time lag between this change and the time the next tests were administered, so this jump in participation rate did not show until 2018. Given that we have combined the Pricetone Review 16-17 academic data (which does incude Colorado and Illunois as states that mandate the SAT) with our SAT/ACT data form 2017 (which does not yet reflect the effect of mandating SAT and ACT in Colorado and Illunois and maintains the old low participation rates prior the mandate occured), its best to exclude Colorado and Illunois. If we don't exclude these states, then this would erroneously skew the results of SAT participation rate in states where SAT is mandatory and free.

### The categories in the barchart:<br/>

**only act and free** - these staes require the act and administer it to the students free of charge (note: ACT participation rate ~100%)<br/>

**either** - either the SAT or the ACT is required based on students' personal choce, but they are not free (note: in this case students tend to choose the ACT a bit more often than the SAT)<br/>

**either and act is free** - these are states that required either the SAT or the ACT based on students' personal choice but administer the ACT tests free of charge to students (note: here, ACT participation is much higher than SAT participation - naturally, given the fact that ACT is free)<br/>

**neiter** - neither the SAT or the ACT are required, and neither one of them are free if the students choose to take them anyway (note: students tend to take ACT and SAT roughly at similar rates)<br/>

**neither but both free** - neither test is required in these states, but should students choose to take them, both tests are free of charge (note: much higher participation rate in ACT in these states. More analyses is needed to determine why - perhaps due to geographic location of the state and relative popularity of the tests in that area)<br/>

**only sat and free** - states in which the SAT is required and is administered free of charge to students (note: SAT participation rate close to 100%)

## 3. Statistical Inference 

### Comments on Statistical Inference to Understand Relationship Between ACT and SAT Participation Rates

A question arises whether it makes sens to do statistical inferencing between the ACT and SAT participation rates. 
Let's say we want to conduct a hypotheses test to compare ACT and SAT participation rates because we note some difference in them and want to determine whether the difference is statistically significant. Well, it makes no sense to do that given that we know from research that the relationship between the two is going to be roughly inversly proportional - causing the polarity between SAT and ACT participation rates we see for a given year in the visualizations above. We know that this difference is expected and observed in the data because of state regulations on mandating either one of the tests. So there is no need to conduct further hypotheses testing. 

Another question arises as to whether it makes sense to compare SAT and ACT math scores. We know that the format and structure of the ACT and SAT are different. The SAT is more commonly known as the 'aptitude' test, while ACT is generally known to test the participants' cumulative knowledge base. So the questions in the math sections of these tests are testing sutdents in different ways and for different things. Furthermore, students are being scored quite differently, with SAT math score range in the hundreds and the ACT score range that goes up to 36. The scales are significantly different and they cannot be easily normalized either for comparison purposes due to the different things they measure and the different approaches of the test administered. Therefore it makes not much sense to compare ACT and SAT math scores and each one should be assessed separately.
